```python
import torch
```

```python
import numpy as np
```

```python
print(torch.__version__)
```

```python
# 1D Tensor
t1 = torch.tensor([1, 2, 3])

# 2D Tensor (Matrix)
t2 = torch.tensor([[1, 2], [3, 4]])

# 3D Tensor
t3 = torch.tensor([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
# printing shapes of the tensors
print(t1.shape, t2.shape, t3.shape)
```

```python
# 1D Array
n1 = np.array([1, 2, 3])

# 2D Array
n2 = np.array([[1, 2], [3, 4]])

# 3D Array
n3 = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
# printing shapes of the numpy
print(n1.shape , n2.shape , n3.shape)
```

```python
# creating two tensors 
a = torch.tensor([10, 20, 30])
b = torch.tensor([1, 2, 3])
# performing + , - , * , / operations 
print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

```python
#creating two array 
a = np.array([10, 20, 30])
b = np.array([1, 2, 3])
# # performing + , - , * , / operations 
print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

```python
x = torch.tensor([1, 2, 3])
y = torch.tensor([4, 5, 6])
# calculating dot product of 2 tensors
print(torch.dot(x, y))
```

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
# calculating dot product of 2 arrays
print(np.dot(x, y))
```

```python
# creating two 2x2 matrices 
A = torch.tensor([[1, 2], [3, 4]])
B = torch.tensor([[5, 6], [7, 8]])
# performing matrix multiplication 
print(torch.matmul(A, B))
# performing matrix multiplication by operator '@'
print("matrix multiplication by using operator '@' ")
print(A @ B)
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
# performing matrix multiplication
print(np.matmul(A, B))
# performing matrix multiplication by operator '@'
print("matrix multiplication by using operator '@' ")
print(A @ B)
```
